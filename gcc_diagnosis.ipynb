# GCC Compilation Diagnosis
This notebook reproduces the GCC failure, captures compiler diagnostics, inspects the environment, and verifies the build output.

## 1. Set Up the C Source File
Create a sample source file from the user's code.

```python
source = '''#include <stdio.h>

int main() {
    printf("Hello world!\n");
    printf("IDE setup test\n");
    printf("my first program\n");
    return 0;
}
'''
with open('01_first.c', 'w', encoding='utf-8') as f:
    f.write(source)
print('Created 01_first.c')
```

## 2. Run GCC with Default Options
Invoke gcc and capture its output.

```python
import subprocess, os
cmd = ['gcc', '01_first.c', '-o', '01_first.exe']
proc = subprocess.run(cmd, capture_output=True, text=True)
print('returncode=', proc.returncode)
print('stdout=')
print(proc.stdout)
print('stderr=')
print(proc.stderr)
```

## 3. Inspect GCC Error Messages
Show compiler diagnostics to identify the root cause.

```python
print('GCC available:', subprocess.run(['where', 'gcc'], capture_output=True, text=True).stdout)
print('GCC version:', subprocess.run(['gcc', '--version'], capture_output=True, text=True).stdout)
print('Environment PATH contains msys64:', 'C:\msys64' in os.environ.get('PATH', ''))
print('Executable exists:', os.path.exists('01_first.exe'))
```

## 4. Fix Syntax and Semantic Errors
The source file is already valid, so this section is a placeholder.

## 5. Compile with Debug Flags
Re-run gcc with additional warnings and debug symbols.

```python
cmd = ['gcc', '-Wall', '-Wextra', '-g', '01_first.c', '-o', '01_first_debug.exe']
proc = subprocess.run(cmd, capture_output=True, text=True)
print('returncode=', proc.returncode)
print('stderr=')
print(proc.stderr)
```

## 6. Check Include Paths and Libraries
Verify GCC search paths and header locations.

```python
proc = subprocess.run(['gcc', '-v', '-E', '01_first.c', '-o', 'nul'], capture_output=True, text=True)
print('returncode=', proc.returncode)
print(proc.stderr)
```

## 7. Validate GCC Installation and Version
Inspect the compiler installation and runtime environment.

```python
print('GCC path:', subprocess.run(['where', 'gcc'], capture_output=True, text=True).stdout)
print('cc1 path:', subprocess.run(['gcc', '-print-prog-name=cc1'], capture_output=True, text=True).stdout)
print('GCC search dirs:', subprocess.run(['gcc', '-print-search-dirs'], capture_output=True, text=True).stdout)
```

## 8. Run the Executable
Run the compiled program if it exists.

```python
if os.path.exists('01_first.exe'):
    proc = subprocess.run(['01_first.exe'], capture_output=True, text=True)
    print('exe returncode=', proc.returncode)
    print(proc.stdout)
    print(proc.stderr)
else:
    print('01_first.exe not found')
```
